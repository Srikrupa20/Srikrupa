# cost function

##  $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

## $$ f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

```python
def f(x,y):
    
    r = 3**(-x**2 - y**2)
    
    return 1/(r+1)
```

## cost function

##  $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

## Partial Derivative of the cost function

## $$\frac{\partial f}{\partial x} = \frac{2x \ln(3) \cdot 3^{-x^2 - y^2}}{\left( 3^{-x^2 - y^2} + 1 \right) ^2}$$

## $$\frac{\partial f}{\partial y} = \frac{2y \ln(3) \cdot 3^{-x^2 - y^2}}{\left( 3^{-x^2 - y^2} + 1 \right) ^2}$$

https://www.symbolab.com/

```python
def fpx(x, y):
    
    r = 3**(-x**2 - y**2)
        
        
    return  2*x*log(3)*r / (r +1)**2 
```

```python
def fpy(x, y):
        
    r = 3**(-x**2 - y**2)
        
    return  2*y*log(3)*r / (r +1)**2 
```

```python
params = np.array([1.8, 1])

max_iter = 1000

multiplier = 0.1


for n in range(max_iter):
    
    gradient_x = fpx(params[0], params[1])
    
    gradient_y = fpy(params[0], params[1])
    
    gradients = np.array([ gradient_x, gradient_y])
    
    params = params - multiplier * gradients
    
```

```python
params = np.array([1.8, 1])

max_iter = 1000

multiplier = 0.1

values_array = params.reshape(1,2)


for n in range(max_iter):
    
    gradient_x = fpx(params[0], params[1])
    
    gradient_y = fpy(params[0], params[1])
    
    gradients = np.array([ gradient_x, gradient_y])
    
    params = params - multiplier * gradients
    
    values_array = np.append(values_array, params.reshape(1,2), axis = 0)
    
```

```python
fig = plt.figure(figsize = (7,7)).add_subplot(projection = '3d')


fig.set_xlabel('X', fontsize = 20, c = 'red')

fig.set_ylabel('Y', fontsize = 20, c = 'red')

fig.set_zlabel('f(X,Y)....Cost Function', fontsize = 20, c = 'red')

fig.plot_surface(x_4, y_4, f(x_4, y_4), alpha = 0.3, cmap = cm.coolwarm)

fig.scatter(values_array[:,0], values_array[:,1], f(values_array[:,0],values_array[:,1]), color = 'red', alpha = 0.3)

plt.show()
```
